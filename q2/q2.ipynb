```python
# Bivariate distribution using seaborn's penguins dataset
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset
penguins = sns.load_dataset("penguins")

# Drop rows with missing values
penguins = penguins.dropna()

# Plot bivariate distribution
plt.figure(figsize=(8, 6))
sns.kdeplot(
    data=penguins, x="bill_length_mm", y="flipper_length_mm",
    cmap="Blues", fill=True, thresh=0.05, levels=50
)
plt.title("Bivariate Distribution of Bill Length and Flipper Length")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Flipper Length (mm)")
plt.show()

```
