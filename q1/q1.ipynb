```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KernelDensity

# Generate synthetic data from a mixture of Gaussians
np.random.seed(42)
data = np.concatenate([
    np.random.normal(loc=-2, scale=1, size=500),
    np.random.normal(loc=3, scale=0.5, size=500)
])

# Define the range for the density estimation
x_range = np.linspace(-6, 6, 1000).reshape(-1, 1)

# Fit KDE with Gaussian kernel
kde = KernelDensity(kernel='gaussian', bandwidth=0.5)
kde.fit(data.reshape(-1, 1))

# Evaluate the density
log_density = kde.score_samples(x_range)
density = np.exp(log_density)

# Plot the results
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, density=True, alpha=0.5, color='gray', label='Histogram')
plt.plot(x_range, density, color='blue', label='KDE (Bandwidth=0.5)')
plt.title('Kernel Density Estimation')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```
